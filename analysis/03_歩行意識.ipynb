```python
df = pd.read_csv('./data/isiki.csv',index_col=0)

## データを絞る
'''
時間軸(年次) : 2010年
性別： 総数
年齢階級 : '20歳-29歳','30歳-39歳','40歳-49歳','50歳-59歳','60歳-69歳'
回答（はい・いいえ）:はい
'''
age_list =  ['20歳-29歳','30歳-39歳','40歳-49歳','50歳-59歳','60歳-69歳']
df = df[(df['時間軸(年次)'] == '2010年') & (df['性別'] == '総数') & (df['年齢階級'].isin(age_list)) & (df['回答（はい・いいえ）'] == 'はい')]
df.to_csv('./data/isiki2.csv')
df
```

```python
df.sort_values('平均値')
```

```python
# 年次別に値と値2の合計を算出
df2 = df.groupby('歩数を増やすことの意識').agg({
    '人数': 'sum',
    '平均値': 'mean'
})
df3 = df2.sort_values('平均値').copy()
df3
```

```python
# 値のフォーマット
format_dict = {'平均値':'{0:,.0f}','人数':'{0:,.0f}'}
# グラフの設定をしながら表示
(df3
 .style
 .format(format_dict)
 .bar(color="#99ceff", vmin=0, subset=['人数'], align='zero')
 .bar(color="#C95A56", vmin=0, subset=['平均値'], align='zero'))
```
